* The columns'neighbourhood_group'and'license'are all null, columns "name" and "host_name" are irrelevant and insignificant to our data analysis,so we could drop them.
* The column'last_review' is date,if there were no reviews for the listing,date simply will not exist,so I choose 'unknown'to fill it.
* The column 'reviews_per_month'we can simply append it with 0 for missing values.

```python
df.drop(['neighbourhood_group','license','name','host_name'], axis=1, inplace=True)
df['last_review'].fillna(value='unknown',inplace = True)
df['reviews_per_month'].fillna(0,inplace = True)
df.isnull().sum()
```
