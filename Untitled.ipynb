Question: What is the balance of hosts each year?

Answer:$$blance(host) = created(host)-left(host)$$

```python
# Merge the two tables where host was created and left
concatenated = pd.concat([df_host, df_host_left], axis="columns")
concatenated.reset_index(inplace=True)
concatenated.columns = concatenated.columns.str.replace('index', 'year')
#fill the null with 0 of host left 
concatenated.fillna(value=0,inplace = True)
# drop the last row of last review in 2022 which is not considering host left
concatenated.drop(index=concatenated.index[-1],axis=0,inplace=True)
# The balance of hosts is equal to the number created minus the number left
concatenated['host_balance_count']= concatenated['host_count']-concatenated['host_left_count']

# Visualizing Data
fig, axs = plt.subplots(figsize=(10,3))
ax =sns.barplot(data=concatenated, x='year' ,y='host_balance_count')
ax.bar_label(ax.containers[0])
plt.title('The balance number of hosts each year')
```

### Next, I will calculate the growth rate of host to estimate the number of hosts if there is no pandemic, so as to assess the specific impact of the pandemic on the number of hosts. First, I will use the monthly growth rate to predict the number of hosts in 2020 and 2021, and then use the quarterly growth rate to reconfirm the forecast.

#### monthly growth rate

Question: What is the growth rate by month from 2009 to 2019 (before pandemic)?

Answer:

```python
# Creat a dataframe of host were created before pandemic
df_host_before_pandemic = df.loc[df['host_since_year'].isin([2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])]
df_host_before_pandemic = df_host_before_pandemic.groupby(['host_since_year','host_since_month'])['id'].count().reset_index(name='created_count')
df_host_before_pandemic.rename(columns={'host_since_year': 'year','host_since_month':'month'}, inplace=True)
# Creat a dataframe of host left before pandemic
df_before_pandemic = df.loc[df['last_review_year'].isin([2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019])]
df_before_pandemic = df_before_pandemic.groupby(['last_review_year','last_review_month'])['id'].count().reset_index(name='left_count')
df_before_pandemic.rename(columns={'last_review_year': 'year','last_review_month':'month'}, inplace=True)
df1=df_host_before_pandemic.set_index(['year','month'])
df2=df_before_pandemic.set_index(['year','month'])
# join two dataframes
Merge = pd.merge(df1, df2, how='outer',left_index=True, right_index=True)
Merge.fillna(0,inplace=True)
Merge['month_balance']=Merge['created_count']-Merge['left_count']
Merge['sum_month_balance']=Merge['month_balance'].cumsum()
# growth rate
Merge['month_growth_rate'] = Merge['sum_month_balance'].pct_change()
Merge=Merge.reset_index()
Merge
```

```python
# Visualizing Data
fig, ax1 = plt.subplots(figsize=(10, 5))
sns.lineplot(data = Merge, x = Merge.index ,y= 'month_growth_rate')
plt.title('The growth rate by month from 2009 to 2019')
```

* From the chart above, we can see that the monthly growth rate fluctuates widely at first and then flattens out.

Question: What is the difference between the number of hosts with and without pandemic?

Answer:
* According to the growth rate before the pandemic, the changes in recent years were relatively stable. Therefore, I selected the average growth rate of 12 months in 2019 to calculate the host count without the pandemic. And then finally I'm going to visualize the result by the amount that comes out.
$$ difference(host) = predict(host)-actual(host)$$

```python
# Create a function to calculate the number of hosts in 2020 and 2021
def predict_count(a,rate, n):# a is the last actual host count in 2019,rate=1+growth rate, n is number of month
    list = [a*rate]
    for i in range(1,n):
        list.append(list[i-1]*rate)
    return(list[-1])
```

```python
# description of the the monthly growth rate for 2019
Merge.loc[Merge['year']==2019]['month_growth_rate'].describe()
```

```python
concatenated['predict_host_count']=concatenated['host_actual_count']
concatenated.loc[11,'predict_host_count']=predict_count(10584,1.002385,12)
concatenated.loc[12,'predict_host_count']=predict_count(10584,1.002385,24)
# predict 2020 and 2021 host count by using monthly growth rate in 2019
# 10584 is the actual host count in the last month of 2019
# 0.002385 is the mean of growth rate of 12 months in 2019
# 24 is 24 months(12 months in 2020 plus 12months in 2021)
concatenated['difference']=concatenated['predict_host_count']-concatenated['host_actual_count']
concatenated['predict_host_count']= concatenated['predict_host_count'].astype(int)
concatenated['difference']=concatenated['difference'].astype(int)
concatenated
```

```python
# Visualizing Data
fig, axs = plt.subplots(figsize=(10,5))
sns.lineplot(data=concatenated, x='year' ,y='host_actual_count',color='green')
sns.lineplot(data=concatenated, x='year',y='predict_host_count',color = 'red')
plt.legend(labels=['host_actual_count', 'predict_host_count'])
plt.title('The actual and predict host count of pandemic')
```

#### quarterly growth rate
Question: What is the growth rate by quarter from 2009 to 2019 (before pandemic)?

Answer:

```python
pd.set_option('display.max_rows', 14)
# host count by Quarter from 2009 to 2022
df_host_Q = df.groupby(df['host_since'].dt.to_period('Q'))['id'].count()
df_host_left_Q = df.groupby(df['last_review'].dt.to_period('Q'))['id'].count()
join= pd.concat([df_host_Q, df_host_left_Q], axis="columns")
join.fillna(0,inplace=True)
join.columns = ['host_created', 'host_left']
join['balance']=join['host_created']-join['host_left']
# drop last 4 quarters in 2022
join.drop(join.tail(4).index,inplace = True)
join = join.reset_index()
join['actual_host_count']=join['balance'].cumsum()
# all quarters from 2009 to 2021
join
```

```python
# 2009-2019 quaterly growth rate
join_2019= join.drop(join.tail(8).index)
join_2019['growth_rate_Q']= join_2019['actual_host_count'].pct_change()

# Draw Plot
plt.figure(figsize=(10,6), dpi= 80)
plt.plot(join_2019.index, 'growth_rate_Q', data=join_2019, color='tab:green')
plt.ylim(-1, 4)
plt.xticks(fontsize=12, horizontalalignment='center', alpha=.7)
plt.yticks(fontsize=12, alpha=.7)
plt.title("Hosts Quarterly growth rate before pandemic (2009 - 2019)", fontsize=20)
plt.grid(axis='both', alpha=.3)
plt.gca().spines["top"].set_alpha(0.0)    
plt.gca().spines["bottom"].set_alpha(0.3)
plt.gca().spines["right"].set_alpha(0.0)    
plt.gca().spines["left"].set_alpha(0.3)   
plt.show()
```

* From the chart above, we can see that the quarterly growth rate trend is same as the monthly growth rate,fluctuates widely at first and then flattens out.
