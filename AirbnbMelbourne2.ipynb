### I think of the date of the last review as approximately the date the host was last active except reviews in 2022. So next, I will calculate the host retention rate based on the last review and the total number of hosts.

```python
df2['last_review']
```

```python
df2['last_review_year'].isnull().sum()
# do not know how to deal with the null
```

```python
# I chose to separate the year and month of the date for the convenience of later analysis.
df2['last_review'] = pd.to_datetime(df2['last_review'])
df2['last_review_year'] = pd.DatetimeIndex(df2['last_review']).year
df2['last_review_month'] = pd.DatetimeIndex(df2['last_review']).month
```

```python
fig, axs = plt.subplots(figsize=(10,3))
ax =sns.countplot(x='last_review_year', data=df2)
ax.bar_label(ax.containers[0])
plt.title('Last review in each year ')
```

```python
df2_last_review = df2.groupby('last_review_year')['last_review_year'].count().reset_index(name='host left')
df2_last_review['sum_left'] = df2_last_review['host left'].cumsum()
df2_last_review
```

```python
df2_total_host = df2_total_host.set_index(['host_since_year'])
df2_last_review = df2_last_review.set_index(['last_review_year'])
concatenated = pd.concat([df2_total_host, df2_last_review], axis="columns")
concatenated
```

```python
concatenated.fillna(value=0,inplace = True)
concatenated['host_active'] = concatenated['sum_count']-concatenated['sum_left']
concatenated['Host_Retention Rate'] = concatenated['host_active']/concatenated['sum_count']*100
concatenated
```

```python
# last review in 2022 not considering left
concatenated.drop(index = 2022,inplace=True)
```

```python
concatenated
```

```python
fig, axs = plt.subplots(figsize=(10,5))
sns.lineplot(data = concatenated, x = concatenated.index ,y= 'sum_count',color = 'green')
sns.lineplot(data = concatenated, x = concatenated.index,y= 'sum_left',color = 'red')
sns.lineplot(data = concatenated, x = concatenated.index ,y= 'host_active',color = 'black')
```

```python
fig, axs = plt.subplots(figsize=(10,5))
sns.lineplot(data = concatenated, x = concatenated.index ,y= 'Host_Retention Rate')
plt.title('Host Retention Rate from 2009 to 2022')
```

### From this graph we can see that while the total number of hosts is increasing every year, the retention rate of hosts is decreasing due to the left (no longer active) of some hosts . Retention rate dropped from 100% to 73%.
### Most of the host left is concentrated after 2018, and it's not hard to imagine covid-19 as the cause. So next, I will divide the time line into three sections: 2009 to 2018(before pandemic),2019 to 2021(pandemic), and finally 2022(after pandemic).
